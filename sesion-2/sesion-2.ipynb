## Crear un archivo CSV

```python
with open('mis_datos.csv', 'w', newline="") as file:
    writer = csv.writer(file)

    writer.writerow(['ID', 'Name', 'Age'])
    writer.writerow(['1', 'Lorena', '25'])
    writer.writerow(['2', 'Saul', '35'])
```
